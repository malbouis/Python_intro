```python
%%writefile Ponto.py
class Ponto:
    """ Cria um novo Ponto, com coordenadas x, y """

    def __init__(self, x=0, y=0):
        """ Inicializa em x, y o novo ponto criado pela classe """
        self.x = x
        self.y = y

    def distancia_da_origem(self):
        """ Calcula minha distânica da origem """
        return ((self.x ** 2) + (self.y ** 2)) ** 0.5

    def distancia_entre_pontos(self, alvo):
        """ Calcula a distânica entre dois pontos """
        dx = self.x - alvo.x
        dy = self.y - alvo.y
        return ((dx ** 2) + (dy ** 2)) ** 0.5

    def reflexao_x(self):
        """ Retorna o ponto refletido no eixo x """
        return Ponto(self.x, -self.y)

    def ponto_medio(self, alvo):
        """ Retorna o ponto medio entre esse ponto e o alvo """
        mx = (self.x + alvo.x)/2
        my = (self.y + alvo.y)/2
        return Ponto(mx, my)

    def parametros_reta(self, alvo):
        """Retorna os parametros da reta determinada por dois pontos"""
        a = abs(self.y - alvo.y)/abs(self.x - alvo.x)
        b = self.y - a*self.x
        return (a,b)

    def inclinacao_origem(self):
        return self.y/self.x

    def __str__(self):
        return "({0}, {1})".format(self.x, self.y)
```

6) Escreva uma definição para uma classe chamada **Circulo** com atributos ***centro*** e ***raio***, em que ***centro*** é um objeto do tipo **Ponto** e ***raio*** é um número.
   * Instancie um objeto **Circulo** que represente um círculo com seu centro em (150, 100) e raio 75 e imprima na tela as coordenadas (x,y) do centro do círculo.
   * Escreva uma função chamada ponto_no_circulo que use um Circulo e um Ponto e retorne True se o Ponto estiver dentro do círculo ou no limite do mesmo.
   


```python
from Ponto import *
class Circulo:
    """Representa um circulo.

    atributos:
    - centro, do tipo Ponto
    - raio, do tipo float
    """
    def __init__(self, centro, raio):
        """ Inicializa os atributos centro e raio """
        self.centro = centro
        self.raio = raio

    def ponto_no_circulo(self, ponto):
        """Verifica se um dado ponto está dentro do círculo.

        ponto: objecto do tipo Ponto
        circulo: objeto do tipo circulo
        """
        d = ponto.distancia_entre_pontos(self.centro)
        print(d)
        return d <= self.raio

    def __str__(self):
        return "({0}, {1})".format(self.centro, self.raio)

```

```python
circulo = Circulo(Ponto(150, 100), 75)
print(circulo)
```

```python
ponto = Ponto(100, 100)
circulo.ponto_no_circulo(ponto)
```
