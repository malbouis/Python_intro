## Ordem das Operações

Quando mais de um operador aparece em uma expressão, a ordem de avaliação depende das **regras de precedência**. Para operadores matemáticos, o **Python segue a convenção matemática**. O acrônimo **PEMDAS (Parênteses, Exponenciação, Multiplicação e Divisão)** é uma maneira útil de lembrar as regras:

**P**arênteses: têm a precedência mais alta e podem ser usados para forçar uma expressão a ser avaliada na ordem desejada. As expressões entre parênteses são avaliadas primeiro.

Você também pode usar parênteses para melhorar a legibilidade, mesmo não alterando o resultado.

```python
4+5*2
```

```python
1/2*2
```

```python
2**2*3
```

```python
2**(2*3)
```

```python
2**(2-2)
```

***Na dúvida, use sempre parênteses!***

# Funções

Em programação, uma função é uma sequência de declarações (à qual atribuímos um nome) que realiza uma tarefa, podendo ou não retornar um valor.

Ao criar uma função, deve-se especificar o nome da função e a sequência de declarações.

## Exemplos de função

```python
type(40)
```

* nome da função: type
* argumento: 40
* valor de retorno da função: é o resultado que a função retorna. Nesse exemplo, a função de nome type, retorna o tipo do argumento passado na função.

```python
print('Olá')
```

* nome da função: print
* argumento: 'Olá'
* valor de retorno da função: é o resultado que a função retorna. Nesse exemplo, a função de nome print, retorna a string que foi usada de argumento.

## Declaração de uma função

Em Python, usamos a palavra-chave **def** para declarar uma função.

```python
def nome_da_funcao():
    # código que será executado
```

Exemplo:

```python
def saudacao():
    print("Olá, seja bem-vindo!")
```

Depois de declarar a função, precisamos chamá-la para que ela execute.

```python
saudacao()
```

Podemos declarar uma função com um ou mais argumentos (também chamados de parâmetros):

```python
def nome_do_aluno(nome):
  print('O nome do aluno é', nome)
```

```python
nome_do_aluno()
```

Podemos declarar uma função com mais de um argumento:

```python
def nome_e_idade(nome, idade):
  print('O monitor', nome, 'tem',idade,'anos')
```

```python
nome_e_idade('Matheus',30)
```

Funções podem retornar um valor

```python
def soma(valor_1,valor_2):
  resultado = valor_1 + valor_2
  return resultado
```

```python
soma(5,3)
```

### Como importar uma função de uma biblioteca do Python

As funções apresentadas até o momento são funções built-in do Python, ou seja, já fazem parte da linguagem e estão disponíveis automaticamente sempre que você instala o Python.

Entretanto, existem muitas outras funções que não estão disponíveis por padrão. Essas funções fazem parte de bibliotecas ou módulos, e por isso precisam ser importadas antes de serem utilizadas no programa. Um exemplo são as funções da biblioteca **math**

```python
import math
```

Para se usar uma função da biblioteca, deve-se usar a seguinte lógica:

nome_do_modulo.nome_da_funcao

Exemplo:


```python
math.sqrt(16)
```

Explorando a biblioteca **math** pode-se encontrar funções muito uteis, por exemplo a que retorna o valor de pi

```python
math.pi
```

## Composição

Até agora, analisamos os elementos de um programa - variáveis, expressões e declarações - isoladamente, sem falar sobre como combiná-los.

Um dos recursos mais úteis das linguagens de programação é a capacidade de usar blocos de construção pequenos e compô-los.

Por exemplo, o argumento de uma função pode ser qualquer tipo de expressão, incluindo operadores aritméticos:

```python
import math
def raiz_quadrada(valor):
  resultado = math.sqrt(valor)
  return resultado
```

```python
raiz_quadrada(3**2 + 4**2)
```
