## Exercícios:
1.  Crie uma função que tome um argumento e imprima o valor e o tipo dele.


2. Crie uma função que calcule e imprima velocidade média de um objeto a partir de uma posição inicial, a final e o tempo transcorrido para um objeto em MRU.


3. Crie uma função que calcule e imprima a velocidade de um objeto a partir da aceleração constante e o tempo (MRUA) (p.ex. queda livre).


4. Crie uma função que calcula o volume de uma esfera a partir do raio da esfera

5. Crie uma função que pede ao usuário 5 valores e ela retorna uma string indicando a média aritmética do conjunto de valores informados.

6. Crie uma função celsius_para_fahrenheit(c), onde é pedido a temperatura em celsius e a função retorna a temperatura em fahrenheit

7. Crie uma função que pergunta ao usuário qual sua idade e retorna o número de dias que ele viveu (aproximadamente)

8. Calcule o seu IMC: IMC = M / A² (com a massa em Kg e a altura em metros). Um valor saudável estara --em geral-- entre 20-25. Um bebê de 6 meses "gorducho" tem 70 cm de "comprimento" e 11 kg de massa, qual o IMC dele?

9. Calcule a velocidade final de um objeto em queda livre a partir de 3 metros de altura (exercício 3). Calcule o tempo que esse objeto demora para cair

10. "Com 8 eletricistas podemos fazer a instalação de uma casa em 3 dias". Crie uma função, usando generalização, que permita deduzir o tempo necessário para completar a instalação com base no número de eletricistas

11. "Uma máquina fabrica 5.000 alfinetes em 2 horas". Crie uma função que usa como argumento o número de alfinetes e retorna o tempo necessário para fabricá-los

12. Escreva uma função que, dado o valor da conta de um restaurante, calcule e exiba a gorjeta do garçom, considerando 10% do valor da conta.

13. Crie um programa em Python que:

Peça ao usuário um número inteiro positivo.
Use um loop para mostrar a tabuada desse número de 1 até 10.
Exiba o resultado no formato:


```
5 x 1 = 5
5 x 2 = 10
...
5 x 10 = 50
```



14. Peça um número e calcule o fatorial dele.

15. Crie uma função que pede um número N e calcula a soma de 1 até N.

16. Crie um programa em Python que mostre um relatório com informações matemáticas de vários números.

Para cada número, o programa deve exibir:

O próprio número
O dobro
O triplo
O quadrado

O resultado deve ser exibido no seguinte formato:


```
Número: 4
Dobro: 8
Triplo: 12
Quadrado: 16
```

Aplique essa função para os primeiros 10 números (ex: 1,2,3,4,5...)
